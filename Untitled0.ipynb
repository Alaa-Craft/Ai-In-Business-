
## 2.1 تطبيق خوارزمية DBSCAN

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) هي خوارزمية تجميع تعتمد على الكثافة. تختلف عن K-Means في أنها لا تتطلب تحديد عدد المجموعات مسبقًا، وبدلاً من ذلك تحدد المجموعات بناءً على كثافة النقاط في الفضاء. يمكنها أيضًا تحديد "نقاط الضوضاء" (outliers) التي لا تنتمي إلى أي مجموعة.

تعتمد DBSCAN على معيارين رئيسيين:
- **`eps` (epsilon):** أقصى مسافة بين عينتين لاعتبار إحداهما من جيران الأخرى.
- **`min_samples`:** العدد الأدنى من العينات في المنطقة المجاورة لعينة معينة (بما في ذلك العينة نفسها) لجعلها "نقطة أساسية" (core point).

سنقوم بتجربة DBSCAN على البيانات المقياسية لدينا.


```python

from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler # للتأكد من أننا نستوردها إذا لم تكن موجودة

# تأكدنا مسبقًا من أن scaled_data جاهز للاستخدام من الخطوات السابقة
# scaled_data هو ناتج scaler.fit_transform(features)

# تجربة قيم مختلفة لـ eps و min_samples
# هذه القيم هي مجرد نقطة بداية، قد تحتاجين إلى تعديلها بناءً على النتائج
# تحديد قيم eps و min_samples يمكن أن يكون تحديًا ويتطلب بعض التجربة
# يمكن استخدام مخطط مسافة k-distance plot للمساعدة في اختيار eps
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_data)

# إضافة تسميات المجموعات الناتجة عن DBSCAN إلى DataFrame الأصلي
df['DBSCAN_Cluster'] = dbscan_labels

# تحقق من عدد المجموعات التي تم العثور عليها وعدد نقاط الضوضاء (-1)
print("DBSCAN Cluster counts:")
print(df['DBSCAN_Cluster'].value_counts())

# لمعرفة عدد المجموعات الفعلية (باستثناء نقاط الضوضاء -1)
n_clusters_dbscan = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
print(f"\nNumber of clusters found by DBSCAN (excluding noise): {n_clusters_dbscan}")


```

## 2.2 تطبيق خوارزمية Agglomerative Hierarchical Clustering

التجميع الهرمي التراكمي (Agglomerative Hierarchical Clustering) هو أسلوب تجميع "من الأسفل إلى الأعلى" (bottom-up). يبدأ كل نقطة بيانات كمجموعة خاصة بها، ثم يدمج أزواجًا من المجموعات الأكثر تشابهًا بشكل متكرر حتى يتم الوصول إلى العدد المطلوب من المجموعات أو يتم تحقيق معيار توقف معين.

الميزة الرئيسية للتجميع الهرمي هي أنه ينشئ بنية شجرية (تسمى المخطط الشجري - Dendrogram) توضح كيف يتم تجميع النقاط والمجموعات. يمكننا استخدام هذا المخطط الشجري للمساعدة في تحديد العدد الأمثل للمجموعات بصريًا.

سنستخدم هنا طريقة "Ward linkage" التي تقلل من تباين المجموعات المجمعة.


```python


from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt # للتأكد من استيراد matplotlib.pyplot

# تحديد عدد المجموعات (يمكننا اختيار نفس العدد 4 لغرض المقارنة الأولية مع K-Means)
# يمكننا أيضًا عدم تحديد n_clusters في البداية واستخدام المخطط الشجري لتحديدها.
agg_clustering = AgglomerativeClustering(n_clusters=4) # يمكن تغيير n_clusters لاحقًا
agg_labels = agg_clustering.fit_predict(scaled_data)

# إضافة تسميات المجموعات الناتجة عن Agglomerative Clustering إلى DataFrame الأصلي
df['Agg_Cluster'] = agg_labels

print("Agglomerative Cluster counts (k=4):")
print(df['Agg_Cluster'].value_counts())

# **إنشاء مخطط شجري (Dendrogram) لتحديد العدد الأمثل للمجموعات بصريًا**
# هذا يمكن أن يكون مفيدًا لاختيار n_clusters إذا لم نكن قد حددناها مسبقًا أو لتأكيدها.
plt.figure(figsize=(18, 9)) # زيادة حجم الرسم لتوضيح أفضل
linked = linkage(scaled_data, method='ward') # 'ward' هي طريقة ربط شائعة تقلل من التباين داخل المجموعات
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=False, # False يجعل المخطط أقل فوضوية لعدد كبير من النقاط
           # truncate_mode='level', p=5, # يمكن استخدام هذا لتقليص المخطط الشجري لعرض المستويات العليا فقط
           # show_contracted=True # لإظهار عدد النقاط في كل عقدة عند التقليص
          )
plt.title('Dendrogram for Agglomerative Hierarchical Clustering')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.grid(True)
plt.show()

print("\nAgglomerative Clustering applied successfully and dendrogram displayed.")
```

```python


import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# تأكدي أن هذه المتغيرات (df, scaled_data) موجودة من الخطوات السابقة في Colab
# إذا لم تكوني متأكدة، يرجى تشغيل كل الخلايا من بداية ملف Colab الخاص بكِ.

# قائمة لتخزين قيم Silhouette Score لكل عدد من التكتلات
silhouette_scores = []

# نطاق عدد التكتلات المراد اختباره (من 2 إلى 10)
# نبدأ من 2 لأن Silhouette Score لا يمكن حسابه لتكتل واحد
range_n_clusters = range(2, 11)

for n_clusters in range_n_clusters:
    # تطبيق Agglomerative Clustering بعدد التكتلات الحالي
    agg_clustering = AgglomerativeClustering(n_clusters=n_clusters)
    agg_labels = agg_clustering.fit_predict(scaled_data)

    # حساب Silhouette Score لهذا التكتل
    score = silhouette_score(scaled_data, agg_labels)
    silhouette_scores.append(score)
    print(f"For n_clusters = {n_clusters}, the Silhouette Score is: {score:.4f}")

# رسم Silhouette Score لكل عدد من التكتلات
plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, silhouette_scores, marker='o', linestyle='-')
plt.title('Silhouette Score for Agglomerative Hierarchical Clustering (Optimal K)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(range_n_clusters)
plt.grid(True)
plt.show()

# تحديد أفضل K بناءً على أعلى Silhouette Score
optimal_k_silhouette = range_n_clusters[np.argmax(silhouette_scores)]
max_silhouette_score = np.max(silhouette_scores)

print(f"\nالعدد الأمثل للتكتلات (k) بناءً على Silhouette Score هو: {optimal_k_silhouette}")
print(f"أعلى Silhouette Score تم تحقيقه: {max_silhouette_score:.4f}")

```
