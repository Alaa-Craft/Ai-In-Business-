```python

from sklearn.cluster import KMeans

wss = []  # ŸÑÿ™ÿÆÿ≤ŸäŸÜ ŸÇŸäŸÖÿ© WSS ŸÑŸÉŸÑ ÿπÿØÿØ ŸÖÿ¨ŸÖŸàÿπÿßÿ™

# ÿ™ÿ¨ÿ±ÿ®ÿ© ŸÖŸÜ 1 ÿ•ŸÑŸâ 10 ŸÖÿ¨ŸÖŸàÿπÿßÿ™
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    wss.append(kmeans.inertia_)  # Ÿà.ÿ≥.ÿ≥ = inertia

# ÿ±ÿ≥ŸÖ Elbow Curve
plt.plot(range(1, 11), wss, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WSS (Inertia)')
plt.grid(True)
plt.show()
```


## 2.1 ÿ™ÿ∑ÿ®ŸäŸÇ ÿÆŸàÿßÿ±ÿ≤ŸÖŸäÿ© DBSCAN

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) ŸáŸä ÿÆŸàÿßÿ±ÿ≤ŸÖŸäÿ© ÿ™ÿ¨ŸÖŸäÿπ ÿ™ÿπÿ™ŸÖÿØ ÿπŸÑŸâ ÿßŸÑŸÉÿ´ÿßŸÅÿ©. ÿ™ÿÆÿ™ŸÑŸÅ ÿπŸÜ K-Means ŸÅŸä ÿ£ŸÜŸáÿß ŸÑÿß ÿ™ÿ™ÿ∑ŸÑÿ® ÿ™ÿ≠ÿØŸäÿØ ÿπÿØÿØ ÿßŸÑŸÖÿ¨ŸÖŸàÿπÿßÿ™ ŸÖÿ≥ÿ®ŸÇŸãÿßÿå Ÿàÿ®ÿØŸÑÿßŸã ŸÖŸÜ ÿ∞ŸÑŸÉ ÿ™ÿ≠ÿØÿØ ÿßŸÑŸÖÿ¨ŸÖŸàÿπÿßÿ™ ÿ®ŸÜÿßÿ°Ÿã ÿπŸÑŸâ ŸÉÿ´ÿßŸÅÿ© ÿßŸÑŸÜŸÇÿßÿ∑ ŸÅŸä ÿßŸÑŸÅÿ∂ÿßÿ°. ŸäŸÖŸÉŸÜŸáÿß ÿ£Ÿäÿ∂Ÿãÿß ÿ™ÿ≠ÿØŸäÿØ "ŸÜŸÇÿßÿ∑ ÿßŸÑÿ∂Ÿàÿ∂ÿßÿ°" (outliers) ÿßŸÑÿ™Ÿä ŸÑÿß ÿ™ŸÜÿ™ŸÖŸä ÿ•ŸÑŸâ ÿ£Ÿä ŸÖÿ¨ŸÖŸàÿπÿ©.

ÿ™ÿπÿ™ŸÖÿØ DBSCAN ÿπŸÑŸâ ŸÖÿπŸäÿßÿ±ŸäŸÜ ÿ±ÿ¶Ÿäÿ≥ŸäŸäŸÜ:
- **`eps` (epsilon):** ÿ£ŸÇÿµŸâ ŸÖÿ≥ÿßŸÅÿ© ÿ®ŸäŸÜ ÿπŸäŸÜÿ™ŸäŸÜ ŸÑÿßÿπÿ™ÿ®ÿßÿ± ÿ•ÿ≠ÿØÿßŸáŸÖÿß ŸÖŸÜ ÿ¨Ÿäÿ±ÿßŸÜ ÿßŸÑÿ£ÿÆÿ±Ÿâ.
- **`min_samples`:** ÿßŸÑÿπÿØÿØ ÿßŸÑÿ£ÿØŸÜŸâ ŸÖŸÜ ÿßŸÑÿπŸäŸÜÿßÿ™ ŸÅŸä ÿßŸÑŸÖŸÜÿ∑ŸÇÿ© ÿßŸÑŸÖÿ¨ÿßŸàÿ±ÿ© ŸÑÿπŸäŸÜÿ© ŸÖÿπŸäŸÜÿ© (ÿ®ŸÖÿß ŸÅŸä ÿ∞ŸÑŸÉ ÿßŸÑÿπŸäŸÜÿ© ŸÜŸÅÿ≥Ÿáÿß) ŸÑÿ¨ÿπŸÑŸáÿß "ŸÜŸÇÿ∑ÿ© ÿ£ÿ≥ÿßÿ≥Ÿäÿ©" (core point).

ÿ≥ŸÜŸÇŸàŸÖ ÿ®ÿ™ÿ¨ÿ±ÿ®ÿ© DBSCAN ÿπŸÑŸâ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑŸÖŸÇŸäÿßÿ≥Ÿäÿ© ŸÑÿØŸäŸÜÿß.


```python

from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler # ŸÑŸÑÿ™ÿ£ŸÉÿØ ŸÖŸÜ ÿ£ŸÜŸÜÿß ŸÜÿ≥ÿ™Ÿàÿ±ÿØŸáÿß ÿ•ÿ∞ÿß ŸÑŸÖ ÿ™ŸÉŸÜ ŸÖŸàÿ¨ŸàÿØÿ©

# ÿ™ÿ£ŸÉÿØŸÜÿß ŸÖÿ≥ÿ®ŸÇŸãÿß ŸÖŸÜ ÿ£ŸÜ scaled_data ÿ¨ÿßŸáÿ≤ ŸÑŸÑÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ŸÖŸÜ ÿßŸÑÿÆÿ∑Ÿàÿßÿ™ ÿßŸÑÿ≥ÿßÿ®ŸÇÿ©
# scaled_data ŸáŸà ŸÜÿßÿ™ÿ¨ scaler.fit_transform(features)

# ÿ™ÿ¨ÿ±ÿ®ÿ© ŸÇŸäŸÖ ŸÖÿÆÿ™ŸÑŸÅÿ© ŸÑŸÄ eps Ÿà min_samples
# Ÿáÿ∞Ÿá ÿßŸÑŸÇŸäŸÖ ŸáŸä ŸÖÿ¨ÿ±ÿØ ŸÜŸÇÿ∑ÿ© ÿ®ÿØÿßŸäÿ©ÿå ŸÇÿØ ÿ™ÿ≠ÿ™ÿßÿ¨ŸäŸÜ ÿ•ŸÑŸâ ÿ™ÿπÿØŸäŸÑŸáÿß ÿ®ŸÜÿßÿ°Ÿã ÿπŸÑŸâ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨
# ÿ™ÿ≠ÿØŸäÿØ ŸÇŸäŸÖ eps Ÿà min_samples ŸäŸÖŸÉŸÜ ÿ£ŸÜ ŸäŸÉŸàŸÜ ÿ™ÿ≠ÿØŸäŸãÿß ŸàŸäÿ™ÿ∑ŸÑÿ® ÿ®ÿπÿ∂ ÿßŸÑÿ™ÿ¨ÿ±ÿ®ÿ©
# ŸäŸÖŸÉŸÜ ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ŸÖÿÆÿ∑ÿ∑ ŸÖÿ≥ÿßŸÅÿ© k-distance plot ŸÑŸÑŸÖÿ≥ÿßÿπÿØÿ© ŸÅŸä ÿßÿÆÿ™Ÿäÿßÿ± eps
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_data)

# ÿ•ÿ∂ÿßŸÅÿ© ÿ™ÿ≥ŸÖŸäÿßÿ™ ÿßŸÑŸÖÿ¨ŸÖŸàÿπÿßÿ™ ÿßŸÑŸÜÿßÿ™ÿ¨ÿ© ÿπŸÜ DBSCAN ÿ•ŸÑŸâ DataFrame ÿßŸÑÿ£ÿµŸÑŸä
df['DBSCAN_Cluster'] = dbscan_labels

# ÿ™ÿ≠ŸÇŸÇ ŸÖŸÜ ÿπÿØÿØ ÿßŸÑŸÖÿ¨ŸÖŸàÿπÿßÿ™ ÿßŸÑÿ™Ÿä ÿ™ŸÖ ÿßŸÑÿπÿ´Ÿàÿ± ÿπŸÑŸäŸáÿß ŸàÿπÿØÿØ ŸÜŸÇÿßÿ∑ ÿßŸÑÿ∂Ÿàÿ∂ÿßÿ° (-1)
print("DBSCAN Cluster counts:")
print(df['DBSCAN_Cluster'].value_counts())

# ŸÑŸÖÿπÿ±ŸÅÿ© ÿπÿØÿØ ÿßŸÑŸÖÿ¨ŸÖŸàÿπÿßÿ™ ÿßŸÑŸÅÿπŸÑŸäÿ© (ÿ®ÿßÿ≥ÿ™ÿ´ŸÜÿßÿ° ŸÜŸÇÿßÿ∑ ÿßŸÑÿ∂Ÿàÿ∂ÿßÿ° -1)
n_clusters_dbscan = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
print(f"\nNumber of clusters found by DBSCAN (excluding noise): {n_clusters_dbscan}")


```

## 2.2 ÿ™ÿ∑ÿ®ŸäŸÇ ÿÆŸàÿßÿ±ÿ≤ŸÖŸäÿ© Agglomerative Hierarchical Clustering

ÿßŸÑÿ™ÿ¨ŸÖŸäÿπ ÿßŸÑŸáÿ±ŸÖŸä ÿßŸÑÿ™ÿ±ÿßŸÉŸÖŸä (Agglomerative Hierarchical Clustering) ŸáŸà ÿ£ÿ≥ŸÑŸàÿ® ÿ™ÿ¨ŸÖŸäÿπ "ŸÖŸÜ ÿßŸÑÿ£ÿ≥ŸÅŸÑ ÿ•ŸÑŸâ ÿßŸÑÿ£ÿπŸÑŸâ" (bottom-up). Ÿäÿ®ÿØÿ£ ŸÉŸÑ ŸÜŸÇÿ∑ÿ© ÿ®ŸäÿßŸÜÿßÿ™ ŸÉŸÖÿ¨ŸÖŸàÿπÿ© ÿÆÿßÿµÿ© ÿ®Ÿáÿßÿå ÿ´ŸÖ ŸäÿØŸÖÿ¨ ÿ£ÿ≤Ÿàÿßÿ¨Ÿãÿß ŸÖŸÜ ÿßŸÑŸÖÿ¨ŸÖŸàÿπÿßÿ™ ÿßŸÑÿ£ŸÉÿ´ÿ± ÿ™ÿ¥ÿßÿ®ŸáŸãÿß ÿ®ÿ¥ŸÉŸÑ ŸÖÿ™ŸÉÿ±ÿ± ÿ≠ÿ™Ÿâ Ÿäÿ™ŸÖ ÿßŸÑŸàÿµŸàŸÑ ÿ•ŸÑŸâ ÿßŸÑÿπÿØÿØ ÿßŸÑŸÖÿ∑ŸÑŸàÿ® ŸÖŸÜ ÿßŸÑŸÖÿ¨ŸÖŸàÿπÿßÿ™ ÿ£Ÿà Ÿäÿ™ŸÖ ÿ™ÿ≠ŸÇŸäŸÇ ŸÖÿπŸäÿßÿ± ÿ™ŸàŸÇŸÅ ŸÖÿπŸäŸÜ.

ÿßŸÑŸÖŸäÿ≤ÿ© ÿßŸÑÿ±ÿ¶Ÿäÿ≥Ÿäÿ© ŸÑŸÑÿ™ÿ¨ŸÖŸäÿπ ÿßŸÑŸáÿ±ŸÖŸä ŸáŸä ÿ£ŸÜŸá ŸäŸÜÿ¥ÿ¶ ÿ®ŸÜŸäÿ© ÿ¥ÿ¨ÿ±Ÿäÿ© (ÿ™ÿ≥ŸÖŸâ ÿßŸÑŸÖÿÆÿ∑ÿ∑ ÿßŸÑÿ¥ÿ¨ÿ±Ÿä - Dendrogram) ÿ™Ÿàÿ∂ÿ≠ ŸÉŸäŸÅ Ÿäÿ™ŸÖ ÿ™ÿ¨ŸÖŸäÿπ ÿßŸÑŸÜŸÇÿßÿ∑ ŸàÿßŸÑŸÖÿ¨ŸÖŸàÿπÿßÿ™. ```python


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# ÿ™ÿ£ŸÉÿØŸä ÿ£ŸÜ Ÿáÿ∞Ÿá ÿßŸÑŸÖÿ™ÿ∫Ÿäÿ±ÿßÿ™ (df, scaled_data) ŸÖŸàÿ¨ŸàÿØÿ© ŸÖŸÜ ÿßŸÑÿÆÿ∑Ÿàÿßÿ™ ÿßŸÑÿ≥ÿßÿ®ŸÇÿ© ŸÅŸä Colab
# ÿ•ÿ∞ÿß ŸÑŸÖ ÿ™ŸÉŸàŸÜŸä ŸÖÿ™ÿ£ŸÉÿØÿ©ÿå Ÿäÿ±ÿ¨Ÿâ ÿ™ÿ¥ÿ∫ŸäŸÑ ŸÉŸÑ ÿßŸÑÿÆŸÑÿßŸäÿß ŸÖŸÜ ÿ®ÿØÿßŸäÿ© ŸÖŸÑŸÅ Colab ÿßŸÑÿÆÿßÿµ ÿ®ŸÉŸê.

# ŸÇÿßÿ¶ŸÖÿ© ŸÑÿ™ÿÆÿ≤ŸäŸÜ ŸÇŸäŸÖ Silhouette Score ŸÑŸÉŸÑ ÿπÿØÿØ ŸÖŸÜ ÿßŸÑÿ™ŸÉÿ™ŸÑÿßÿ™
silhouette_scores = []

# ŸÜÿ∑ÿßŸÇ ÿπÿØÿØ ÿßŸÑÿ™ŸÉÿ™ŸÑÿßÿ™ ÿßŸÑŸÖÿ±ÿßÿØ ÿßÿÆÿ™ÿ®ÿßÿ±Ÿá (ŸÖŸÜ 2 ÿ•ŸÑŸâ 10)
# ŸÜÿ®ÿØÿ£ ŸÖŸÜ 2 ŸÑÿ£ŸÜ Silhouette Score ŸÑÿß ŸäŸÖŸÉŸÜ ÿ≠ÿ≥ÿßÿ®Ÿá ŸÑÿ™ŸÉÿ™ŸÑ Ÿàÿßÿ≠ÿØ
range_n_clusters = range(2, 11)

for n_clusters in range_n_clusters:
    # ÿ™ÿ∑ÿ®ŸäŸÇ Agglomerative Clustering ÿ®ÿπÿØÿØ ÿßŸÑÿ™ŸÉÿ™ŸÑÿßÿ™ ÿßŸÑÿ≠ÿßŸÑŸä
    agg_clustering = AgglomerativeClustering(n_clusters=n_clusters)
    agg_labels = agg_clustering.fit_predict(scaled_data)

    # ÿ≠ÿ≥ÿßÿ® Silhouette Score ŸÑŸáÿ∞ÿß ÿßŸÑÿ™ŸÉÿ™ŸÑ
    score = silhouette_score(scaled_data, agg_labels)
    silhouette_scores.append(score)
    print(f"For n_clusters = {n_clusters}, the Silhouette Score is: {score:.4f}")

# ÿ±ÿ≥ŸÖ Silhouette Score ŸÑŸÉŸÑ ÿπÿØÿØ ŸÖŸÜ ÿßŸÑÿ™ŸÉÿ™ŸÑÿßÿ™
plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, silhouette_scores, marker='o', linestyle='-')
plt.title('Silhouette Score for Agglomerative Hierarchical Clustering (Optimal K)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(range_n_clusters)
plt.grid(True)
plt.show()

# ÿ™ÿ≠ÿØŸäÿØ ÿ£ŸÅÿ∂ŸÑ K ÿ®ŸÜÿßÿ°Ÿã ÿπŸÑŸâ ÿ£ÿπŸÑŸâ Silhouette Score
optimal_k_silhouette = range_n_clusters[np.argmax(silhouette_scores)]
max_silhouette_score = np.max(silhouette_scores)

print(f"\nÿßŸÑÿπÿØÿØ ÿßŸÑÿ£ŸÖÿ´ŸÑ ŸÑŸÑÿ™ŸÉÿ™ŸÑÿßÿ™ (k) ÿ®ŸÜÿßÿ°Ÿã ÿπŸÑŸâ Silhouette Score ŸáŸà: {optimal_k_silhouette}")
print(f"ÿ£ÿπŸÑŸâ Silhouette Score ÿ™ŸÖ ÿ™ÿ≠ŸÇŸäŸÇŸá: {max_silhouette_score:.4f}")

```
